```python
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import numpy as np
import pandas as pd
```

```python
# sns.set_style('darkgrid')  # darkgrid, white grid, dark, white and ticks
plt.rc('axes', titlesize=18)  # fontsize of the axes title
plt.rc('axes', labelsize=18)  # fontsize of the x and y labels
plt.rc('xtick', labelsize=18)  # fontsize of the tick labels
plt.rc('ytick', labelsize=18)  # fontsize of the tick labels
plt.rc('legend', fontsize=18)  # legend fontsize
plt.rc('font', size=18)  # controls default text sizes
# plt.figure(figsize=(10, 6), tight_layout=True)
```

```python
def unpack_data(data):
    n_exemplar_list = data["n_exemplar_list"]
    y_list=data["y_list"]
    datasetname = data["datasetname"]
    n_exemplar_max = data["n_exemplar_max"]
    n_sample = data["n_sample"]
    aggregation_mode = data["aggregation_mode"]
    extra_info=data["extra_info"]
    return n_exemplar_list[1:],y_list[1:]
```

```python
vg_noun_data_path="../data/dumped_plot_data_replication/vg_noun_bert_visual_20_100.pkl"
vg_noun_data=unpack_data(pickle.load(open(vg_noun_data_path,"rb")))

vg_verb_data_path="../data/dumped_plot_data_replication/vg_verb_bert_visual_20_100.pkl"
vg_verb_data=unpack_data(pickle.load(open(vg_verb_data_path,"rb")))

df1 = pd.DataFrame(np.array(vg_noun_data).transpose(), columns = ['x', 'y'])
df1['word_type']="VG Nouns"

df2 = pd.DataFrame(np.array(vg_verb_data).transpose(), columns = ['x', 'y'])
df2['word_type']="VG Verbs"

mit_data_path="../data/dumped_plot_data_replication/mit_bert_visual_20_100.pkl"
mit_data=unpack_data(pickle.load(open(mit_data_path,"rb")))

df_mit = pd.DataFrame(np.array(mit_data).transpose(), columns = ['x', 'y'])
df_mit['word_type']="MiT Verbs"

df_visual=pd.concat([df1,df2,df_mit])
```

```python
fig, ax = plt.subplots(1, figsize=(10, 6))
sns.pointplot(x="x",y="y",data=df_visual,hue="word_type",palette=[sns.color_palette("Paired")[7],sns.color_palette("Paired")[3],sns.color_palette("Paired")[-1]],ax=ax)

# Remove the top and right spines from plot
sns.despine()

ax.legend(loc=4)
ax.set_xlim(-1, 21)
ax.set_xticks([0,5,10,15,20])
ax.set_xticklabels([0,5,10,15,20])
ax.set_ylim(0.7, 1.02)
ax.set_yticks([0.7,0.8,0.9,1.0])

# ax.set_xlabel("Number of Exemplars Aggregated",fontsize=18)
# ax.set_ylabel("Relative Alignment Strength", fontsize=20)

ax.set_xlabel("",fontsize=18)
ax.set_ylabel("", fontsize=20)
# ax.get_legend().remove()

# ax.text(5,1.07,"Visual Aggregation")

plt.tight_layout()
plt.savefig("../figs/visual_agg.png",dpi=600)
plt.show()
```
