```python
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import numpy as np
import pandas as pd
```

```python
def unpack_data_2D(data):
    y_mat=data["y_mat"]
    datasetname = data["datasetname"]
    n_l_exemplar_max = data["n_l_exemplar_max"]
    n_v_exemplar_max = data["n_v_exemplar_max"]
    n_sample = data["n_sample"]
    aggregation_mode = data["aggregation_mode"]
    extra_info=data["extra_info"]
    return y_mat[:10,:10]
```

```python
vg_noun_data_path="../data/dumped_plot_data/2D_vg_noun_bert_visual_language_20_20_1.pkl"
vg_noun_data=unpack_data_2D(pickle.load(open(vg_noun_data_path,"rb")))
```

```python
vg_verb_data_path="../data/dumped_plot_data/2D_vg_verb_bert_visual_language_20_20_1.pkl"
vg_verb_data=unpack_data_2D(pickle.load(open(vg_verb_data_path,"rb")))
```

```python
plt.figure(figsize=(10, 8))
ax = sns.heatmap(vg_noun_data,vmin=0,vmax=1,annot=True,fmt=".3f",cmap="YlGnBu",xticklabels=range(1,11),yticklabels=range(1,11))
ax.invert_yaxis()
plt.xlabel("Number of Language Exemplars Aggregated",fontsize=18)
plt.ylabel("Number of Visual Exemplars Aggregated",fontsize=18)
plt.tight_layout()
plt.savefig("../figs/vg_noun_bert_heatmap.png")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
ax = sns.heatmap(vg_verb_data,vmin=0,vmax=1,annot=True,fmt=".3f",cmap="YlGnBu",xticklabels=range(1,11),yticklabels=range(1,11))
ax.invert_yaxis()
plt.xlabel("Number of Language Exemplars Aggregated",fontsize=18)
plt.ylabel("Number of Visual Exemplars Aggregated",fontsize=18)
plt.tight_layout()
plt.savefig("../figs/vg_verb_bert_heatmap.png")
plt.show()
```
